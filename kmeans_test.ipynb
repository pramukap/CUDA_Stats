```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from clustering import KMeans
```

```python
data = np.array(pd.read_csv('lcluster.csv',header=None))
data
```

```python
model = KMeans(k=15, iterations=1000)
```

```python
ks = model.fit(data)
ks
```

```python
labels = model.label(data)
```

```python
def scatter_classes(data,labels,k,x,y):
    for ki in range(k):
        plt.scatter(data[[labels==ki]][:,x],data[[labels==ki]][:,y],alpha=0.1)
```

```python
scatter_classes(data,labels,15,0,1)
```

```python
label, counts = np.unique(labels, return_counts=True)
print(counts)
```

```python
def distance(X,Y):
    return np.sum([(x-y)**2 for x,y in zip(X,Y)])
```

```python
ks[0]
```

```python
print(distance(ks[0],data[0]),
distance(ks[1],data[0]),
distance(ks[2],data[0]))
```

```python
def get_arg_min_dist(vector,ks):
    return np.argmin([distance(vector,ksi) for ksi in ks])

get_arg_min_dist(data[0],ks)
```
